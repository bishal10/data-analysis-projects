# Google Search Analysis with python

#### Importing all necessary libraries

```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

#### Create Dataframe for Top 10 countries which search for data analysis

```python
trends.build_payload(kw_list=["Data Analysis"])
data = trends.interest_by_region()
data = data.sort_values(by="Data Analysis", ascending=False)
data = data.head(10)
print(data)
```

#### Plot result in bar graph for better understanding

```python
data.reset_index().plot(x="geoName", y="Data Analysis", 
                        figsize=(20,15), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

#### Put 'Data Analysis' searching trend to data frame
Let's check trend of 'Data Analysis' total search

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Data Analysis'])
data = data.interest_over_time()
```

#### Line Graph shows the google search trend on "Data Analysis"

```python
fig, ax = plt.subplots(figsize=(20, 15))
data['Data Analysis'].plot()
plt.style.use('tableau-colorblind10')
plt.title('Google Search trend for "Data Analysis"', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

By looking at the above graph we can clearly see that every end of the year search trend decrease drasticaly where as every begenning of year search trend increase. And also it tells us search trend is shightly increasing every year.
